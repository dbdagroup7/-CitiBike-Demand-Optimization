```python
from pyspark.sql.functions import udf, col
from pyspark.sql.types import BooleanType

final_table = spark.table("sample_citibike.citibike.finaltable_v2_copy")
```

```python
from pyspark.sql.functions import when, col

final_table = final_table.withColumn(
    "prcp_category",
    when(col("PRCP") == 0, "no_rain")
    .when((col("PRCP") > 0) & (col("PRCP") <= 0.1), "light_rain")
    .when((col("PRCP") > 0.1) & (col("PRCP") <= 0.5), "moderate_rain")
    .otherwise("heavy_rain")
)
```

```python
display(
    final_table
    .select("PRCP", "prcp_category")
    .distinct()
    .orderBy("PRCP")
)
```

```python
from pyspark.sql.functions import expr, to_date, date_format, col

final_table = final_table \
    .withColumn("start_ts", expr("timestamp(started_at)")) \
    .withColumn("end_ts", expr("timestamp(ended_at)")) \
    .withColumn("start_date", to_date(col("start_ts"))) \
    .withColumn("start_time", date_format(col("start_ts"), "HH:mm:ss")) \
    .withColumn("end_date", to_date(col("end_ts"))) \
    .withColumn("end_time", date_format(col("end_ts"), "HH:mm:ss")) \
    .drop("start_ts", "end_ts")
```

```python
display(final_table.limit(10))
```

```python
from pyspark.sql.functions import when, col

final_table = final_table.withColumn(
    "temp_category",
    when(col("TEMP") < 40, "cold")
    .when((col("TEMP") >= 40) & (col("TEMP") <= 70), "moderate")
    .otherwise("hot")
)
```

```python
display(
    final_table
    .select("TEMP", "temp_category")
    .distinct()
    .orderBy("TEMP")
)
```

extract hour from start_time

```python
from pyspark.sql.functions import hour, to_timestamp

final_table = final_table.withColumn(
    "start_hour",
    hour(to_timestamp(col("start_time"), "HH:mm:ss"))
)
```

apply rush-hour logic

```python
final_table = final_table.withColumn(
    "is_rush_hour",
    when(
        ((col("start_hour") >= 7) & (col("start_hour") <= 9)) |
        ((col("start_hour") >= 16) & (col("start_hour") <= 19)),
        True
    ).otherwise(False)
)
```

```python
display(
    final_table
    .select("start_date", "start_time", "weekend", "is_rush_hour")
    .limit(20)
)
```

Extract day_of_week

```python
from pyspark.sql.functions import date_format, col

final_table = final_table.withColumn(
    "day_of_week",
    date_format(col("start_date"), "EEEE")
)
```

Extract month

```python
final_table = final_table.withColumn(
    "month",
    date_format(col("start_date"), "MMMM")
)
```

Extract year

```python
from pyspark.sql.functions import year

final_table = final_table.withColumn(
    "year",
    year(col("start_date"))
)
```

Create season

```python
from pyspark.sql.functions import month, when

final_table = final_table.withColumn(
    "season",
    when(month(col("start_date")).isin(12, 1, 2), "Winter")
    .when(month(col("start_date")).isin(3, 4, 5), "Spring")
    .when(month(col("start_date")).isin(6, 7, 8), "Summer")
    .otherwise("Fall")
)
```

```python
display(
    final_table
    .select("start_date", "day_of_week", "month", "year", "season")
    .distinct()
    .orderBy("start_date")
    .limit(20)
)
```

`trip_duration_minutes Calculation`

```python
from pyspark.sql.functions import expr, col
final_table = final_table \
    .withColumn("start_ts", expr("timestamp(started_at)")) \
    .withColumn("end_ts", expr("timestamp(ended_at)"))
```

Calculate duration in seconds → minutes

```python
from pyspark.sql.functions import unix_timestamp
final_table = final_table.withColumn(
    "trip_duration_minutes",
    (unix_timestamp(col("end_ts")) - unix_timestamp(col("start_ts"))) / 60
)
```

```python
display(final_table.limit(10))
```

```python
from pyspark.sql.functions import round, col

final_table = final_table.withColumn(
    "trip_duration_minutes",
    round(col("trip_duration_minutes"), 2)
)
```

Apply safety checks

```python
from pyspark.sql.functions import when
final_table = final_table.withColumn(
    "trip_duration_minutes",
    when(col("start_ts").isNull() | col("end_ts").isNull(), None)
    .when(col("end_ts") < col("start_ts"), None)
    .otherwise(col("trip_duration_minutes"))
)
```

Optional: cap unrealistic trips --> 
CitiBike trips longer than 24 hours are usually errors.

```python
final_table = final_table.withColumn(
    "trip_duration_minutes",
    when(col("trip_duration_minutes") > 1440, None)
    .otherwise(col("trip_duration_minutes"))
)
```

```python
final_table = final_table.drop("start_ts", "end_ts")
```

```python
display(
    final_table
    .select("started_at", "ended_at", "trip_duration_minutes")
    .limit(20)
)
```

```python
# Rain
final_table = final_table.withColumn(
    "has_rain",
    when(substring(col("FRSHTT"), 2, 1) == "1", True).otherwise(False)
)
```

```python
# Snow 
final_table = final_table.withColumn(
    "has_snow",
    when(substring(col("FRSHTT"), 3, 1) == "1", True).otherwise(False)
)
```

```python
# Thunder
final_table = final_table.withColumn(
    "has_thunder",
    when(substring(col("FRSHTT"), 5, 1) == "1", True).otherwise(False)
)
```

```python
display(
    final_table
    .select("FRSHTT", "has_fog", "has_rain", "has_snow", "has_thunder")
    .distinct()
    .limit(20)
)
```

```python
from pyspark.sql.functions import udf, col
from pyspark.sql.types import BooleanType

final_table1 = spark.table("sample_citibike.citibike.fact_citibike_features")
```

creating is_round_trip column 

```python
from pyspark.sql.functions import when
final_table1 = final_table1.withColumn(
    "is_round_trip",
    when(
        (col("start_station_id") == col("end_station_id")),
        True
    ).otherwise(False)
)
```

Nullify distance only for same-station trips

```python
from pyspark.sql.functions import col, when

final_table1 = final_table1.withColumn(
    "trip_distance",
    when(
        col("is_round_trip") == True,
        None
    ).otherwise(col("trip_distance"))
)
```

validate

```python
final_table1.selectExpr(
    "avg(trip_distance) as avg_distance",
    "count(*) as total_trips",
    "sum(case when trip_distance is null then 1 else 0 end) as null_dist_trips"
).show()
```

```python
final_table1.filter(col("trip_distance") > 1000) \
           .select(
               "trip_distance",
               "trip_duration_minutes",
               "start_lat", "start_lng",
               "end_lat", "end_lng",
               "end_station_id",
               "end_station_name"
           ) \
           .show(truncate=False)
```

```python
from pyspark.sql.functions import col

final_table1 = final_table1.filter(
    (col("trip_distance").isNull()) | (col("trip_distance") <= 1000)
)
```

Deleting the null values from the trip_duration_minutes column

```python
final_table1 = final_table1.filter(
    col("trip_duration_minutes").isNotNull()
)
```

```python
from pyspark.sql.functions import when, col

final_table1 = final_table1.withColumn(
    "wind_category",
    when(col("WDSP") <= 2.5, "low")
    .when((col("WDSP") > 2.5) & (col("WDSP") <= 4.5), "moderate")
    .otherwise("high")
)
```
